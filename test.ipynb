```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

import requests
import time
```

```python
def parent(element):
    return element.find_element(by=By.XPATH, value='./..')
```

```python
driver = webdriver.Chrome("D:/Work/NECTEC internship/Sentiment/code/chromedriver.exe")
```

```python
driver.get('https://tiktok.com/')
```

```python
pre = -1
video_containers = []
index = 0
limit = 3

record = ""

while(len(video_containers) > pre):
    video_containers = driver.find_elements(by=By.XPATH, value="//div[@class='tiktok-1p48f7x-DivItemContainer etvrc4k0']")
    
    for i in range(index, min(limit, len(video_containers))):
        container = video_containers[i]
        
        try:
            author = container.find_element(by=By.XPATH, value=".//h3[@class='tiktok-debnpy-H3AuthorTitle emt6k1z0']")
            author_string = author.text.strip()
        except NoSuchElementException:
            author_string = ""
            
        labels = container.find_elements(by=By.XPATH, value=".//span[@class='tiktok-j2a19r-SpanText efbd9f0']")
        label_text = [label.text.strip() for label in labels]
        label_string = "".join(label_text)
        
        tags = container.find_elements(by=By.XPATH, value=".//strong[@class='tiktok-f9vo34-StrongText ejg0rhn1']")
        tag_text = [tag.text.strip() for tag in tags]
        tag_string = ", ".join(tag_text)
        
        try:
            video = container.find_element(by=By.TAG_NAME, value="video")
            video_url = video.get_attribute('src')
            
            response = requests.get(video_url)
            #open(f"results/video {i}.mp4", "wb").write(response.content)
            
            rec = author_string + ", " + label_string + ", " +  tag_string + "\n"
            record += rec
            
            print(rec + ": " + video_url[:50], end='\n')
        except:
            print("Get video failed", end='\n')
        
        size = container.size
        driver.execute_script(f"scrollBy(0, {size['height']});")
        time.sleep(1)
    index = len(video_containers)
    if (index >= limit):
        break
    
    time.sleep(5)

# f = open("results/details.csv", "w")
# f.write(record)
# f.close()
```
